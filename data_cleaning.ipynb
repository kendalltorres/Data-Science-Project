```python
# Columnas que tienen valores NaN
missing_values = df.isnull().sum()
missing_columns = missing_values[missing_values > 0]
print(missing_columns)
```

```python
# Eliminar filas donde 'GarageYrBlt' es NaN ya que el tener o no garage
# es una caracteristica importante y tener filas sin esta informacion
# puede distorsionar los resultados

df = df.dropna(subset=['GarageYrBlt'])
```

```python
# Rellenar NaNs con el promedio ya que aquí los valores son menos críticos,
# podrían faltar simplemente debido a errores de entrada o falta de registro.
# Se puede rellenar con el promedio con el fin de mantener la continuidad y utilidad del conjunto de datos

df['LotFrontage'] = df['LotFrontage'].fillna(df['LotFrontage'].mean())
```

```python
# Conocer la distribucion de las propiedades
unique_neighborhood= df['Neighborhood'].unique()
unique_neighborhood
```

```python
# Un conteo de casas en venta por barrio
neighborhood_counts= df['Neighborhood'].value_counts()
neighborhood_counts
```

```python
# Reemplazar los NaN con la moda (el valor más frecuente) de la columna
# El número de valores faltantes es muy pequeño comparado con el tamaño del dataset,
# por lo que reemplazar con la moda mantendrá la distribución de la variable.
df['MSZoning'] = df['MSZoning'].fillna(df['MSZoning'].mode()[0])
```

```python
# Más del 88% de los valores están faltantes, lo que sugiere que esta columna no tiene suficiente información útil.
df.drop(columns=['Alley'], inplace=True)
```

```python
# Reemplazar los NaN con la moda
df['Utilities'] = df['Utilities'].fillna(df['Utilities'].mode()[0])
```

```python
# Remplazar los 825 valores NaN por None suponiendo que no tienen revestimiento de mamposteria y asi aprovechar los demas datos
df['MasVnrType']  = df['MasVnrType'].fillna('None')

# Cambiar los 14 NaN por 0, ya que es un porcentaje muy pequeño de valores, suponoendo que esos NaN son las casas que no tienen revestimiento
df['MasVnrArea'] = df['MasVnrArea'].fillna(0)
```

```python
# Reemplazar NaN con 'No' para indicar que no hay sótano
# Los valores faltantes probablemente indican la ausencia de un sótano
bsmt_cols = ['BsmtQual', 'BsmtCond', 'BsmtExposure', 'BsmtFinType1', 'BsmtFinType2']
for col in bsmt_cols:
    df[col] = df[col].fillna('No')
```

```python
# Reemplazar NaN con 0
# La ausencia de datos probablemente indica que no hay baños en el sótano
df['BsmtFullBath'] = df['BsmtFullBath'].fillna(0)
df['BsmtHalfBath'] = df['BsmtHalfBath'].fillna(0)
```

```python
# Reemplazar NaN con 0. Un solo valor faltante puede ser tratado como una ausencia de espacio de sótano terminado.

bsmt_sf_cols = ['BsmtFinSF1', 'BsmtFinSF2', 'BsmtUnfSF', 'TotalBsmtSF']
for col in bsmt_sf_cols:
    df[col] = df[col].fillna(0)
```

```python
# Reemplazar el unico NaN con la moda

df['KitchenQual'] = df['KitchenQual'].fillna(df['KitchenQual'].mode()[0])
```

```python
 # Reemplazar los 2 NaN con la moda

df['Functional'] = df['Functional'].fillna(df['Functional'].mode()[0])
```

```python
#  Reemplazar NaN con 'No'
# La ausencia de datos probablemente indica que no hay chimenea

df['FireplaceQu'] = df['FireplaceQu'].fillna('No')
```

```python
# Reemplazar NaN con 'No'.
# Suponiendo que la gran mayoría de las casas no tienen piscina

df['PoolQC'] = df['PoolQC'].fillna('No')
```

```python
# Reemplazar los 1104 valores NaN con 'No'
# La ausencia de esos datos puede indicar que no hay cerca

df['Fence'] = df['Fence'].fillna('No')
```

```python
# Reemplazar NaN con 'No'. Los 1335 valores NaN pueden indicar que no hay caracteristicas adicionales

df['MiscFeature'] = df['MiscFeature'].fillna('No')
```

```python
# Reemplazar NaN con la moda. Solo un valor faltante

df['SaleType'] = df['SaleType'].fillna(df['SaleType'].mode()[0])
```

```python
#Limpiamos el dataset de valores NaN
missing_values = df.isnull().sum()
missing_columns = missing_values[missing_values > 0]
missing_columns
```
